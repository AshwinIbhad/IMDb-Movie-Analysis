# Data Analysis

```python
# Divide the'gross' and 'budget' columns by 1000000 to convert '$' to 'million $'

movies['Gross'] = movies['Gross'].floordiv(1000000)
movies['budget'] = movies['budget'].floordiv(1000000)
```

```python
# Creating the new column named 'profit' by subtracting the 'budget' column from the 'Gross' column

movies.insert(4, column='Profit', value = movies['Gross']-movies['budget'])
```

```python
# Sorting the dataframe with the 'profit' column as reference using the 'sort_values' function and making sure to set
# the argument 'ascending' to 'False'

movies.sort_values(by=['Profit'], ascending = False, inplace = True, ignore_index = True)

# Getting the top 10 profitable movies by using position based indexing

top_10 = movies.loc[0:9, 'Title']
top_10
```

```python
# Setting the theme

sns.set_style('whitegrid')

# Plot profit vs budget

plt.figure(figsize=[8, 6])
h = sns.jointplot(x = 'budget', y = 'Profit', data = movies, color = 'red')
h.set_axis_labels('budget', 'Profit', fontsize = 15)
plt.show()
```

```python
# Find the movies with negative profits

negative_profit = movies[movies['Profit'] < 0]
negative_profit.reset_index(drop = True)

# Getting the top 5 movies that had incurred losses the most

negative_profit.sort_values(by = 'Profit')[0:5]
```

```python
# Adding a column Popularity that sums up the facebook likes of all the three actors

movies.insert(11, column='Popularity', value = movies.loc[:, ['actor_1_facebook_likes',
                                                              'actor_2_facebook_likes',
                                                              'actor_3_facebook_likes']].sum(axis = 1))
```

```python
top_popular_actors = movies.sort_values(by = 'Popularity', ascending = False, ignore_index = True).loc[0:4, :]
top_popular_actors
```
